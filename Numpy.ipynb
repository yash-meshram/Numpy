## Matrix multiplication

```python
a = np.ones((2,3))
print(a)

b = np.full((3,2),2)
print(b)

c = np.matmul(a,b)
print(c)
```

## Matrix determinant

```python
print(np.linalg.det(c))

c = np.identity(3)
print(np.linalg.det(c))
```
